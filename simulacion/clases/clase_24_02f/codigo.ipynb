```python
def funcion(x):
    resul = []
    for i in x: 
        if(0.5<i<1): resul.append(2.0)
        else: resul.append(1.0)

    return np.array(resul)


tiempos = np.linspace(0, 2, 41)
print(tiempos)
resultados = funcion(tiempos)

plt.plot(tiempos, resultados)

plt.show()
```
