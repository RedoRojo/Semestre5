```python
data_cop = np.loadtxt("cooperativa.csv", delimiter=',')
data_cop.shape
```

```python
def diagrama(data_cop):
    precios_cierre = {}
    for fila in data_cop:
        codigo = fila[0]
        dia = fila[1]
        cierre = fila[5]
        if codigo not in precios_cierre: 
            precios_cierre[codigo] = []
        
        precios_cierre[codigo].append([dia, cierre])

    for codigo, cierres in precios_cierre.items(): 
        formato = np.array(cierres)
        dias = formato[:, 0]
        precio_cierre = formato[:, 1]
        plt.plot(dias, precio_cierre)
        
    

diagrama(data_cop)
plt.show()
```
